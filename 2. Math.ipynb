### 1. Support vector machine (SVM)

![SVM - Margin](https://d2h0cx97tjks2p.cloudfront.net/blogs/wp-content/uploads/sites/2/2019/07/introduction-to-SVM.png)

(Utilité : utiliser la descente de gradient.)

#### Type 

Apprentissage supervisé.

#### Utilité ?

- Classification, Régression, Groupement (clustering)

#### Quand l'utiliser ? 

- Lorsqu'on a un nombre relativement faible de données $(\leq 1000)$.

#### Principe

Principe : on a 2 classes de données, si on place ces données sur un graphique, on peut tracer une droite qui sera la frontière de décision qui séparera les 2 classes de données.    

Cette ligne est appelée un **hyperplan**.     
On construit cet hyperplan en maximisant la marge qui est l'espace entre l'hyperplan et les 2 classes ou plus précisément les points les plus proches de l'hyperplan.    
On appelle ces points **"Support Vectors"**


#### Qu'est-ce qu'un hyperplan ?

Un hyperplan est une "surface de décision" qui coupe l'espace en 2.    
Il s'agit d'un **classificateur binaire**.

Si nos données ont $N$ caractéristiques, _l'espace vectoriel_ sera de dimension $N$ et _l'hyperplan_ sera de dimension $N-1$.

Dans $\mathbb{R}^2$, il s'agit d'une droite.    
Dans $\mathbb{R}^3$, il s'agit d'un plan. 

#### Cas non-linéaire ?

![kernel trick](https://miro.medium.com/max/1744/1*zWzeMGyCc7KvGD9X8lwlnQ.png)

Par définition un **hyperplan** est linéaire.
Que faire dans le cas où la ligne de séparation entre nos 2 classes n'est pas linéaire (ex.: une courbe) ? 

**Trick** : Il est possible de créer un espace où cette ligne de séparation devient un hyperplan _(kernel trick)_.

#### Loss function (Hinge loss)

$$ 
c(x, y, f(x)) = (1 - y \cdot f(x))_+ := 
    \begin{cases} 
        0, \quad \textit{si $c(x,y,f(x))$ $\leq$ 0} \\
        1 - y \cdot f(x), \quad \textit{sinon} 
    \end{cases}
$$

- $c$ : "loss function".    
- $x$ : la donnée.    
- $y$ : le véritable label.    
- $f(x)$: le label prédit. 

#### Objective function

On va faire une descente de gradient.
Cela consiste à prendre le minimum selon $w$ (a.k.a. dérivée partielle p/r à $w$) de la fonction suivante en supposant que nous avons $n$ données : 

$$ 
\lambda ||w||^2 + \sum_{i = 1}^n (1 - y_i <x_i | w>))_+
$$

- $w$ : un vecteur

On minimise la fonction : 

$$ 
\frac{\partial}{\partial w}\lambda ||w||^2 |_{w = w_k} = 2 \lambda w_k
$$ 

$$
\frac{\partial}{\partial w} (1 - y_i <x_i | w>))_+ |_{w = w_k} := 
    \begin{cases}
        0, \quad \textit{si $y_i <x_i | w>) \geq 1$} \\
        - y_i x_{ik}, \quad \textit{sinon}
    \end{cases}
$$
